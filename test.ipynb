```python
import numpy as np
from sthrd_benchmark import STHRDBenchmark, AbstractOptimizer, BenchmarkVisualizer

class MySGD(AbstractOptimizer):
    def step(self, x, oracle_data):
        grad = oracle_data.get('grad')
        lr = self.hp.get('lr', 0.01)
        if 'm' not in self.state:
            self.state['m'] = np.zeros_like(x)
        self.state['m'] = 0.9 * self.state['m'] + 0.1 * grad
        return x - lr * self.state['m']

benchmark = STHRDBenchmark()

benchmark.generate_dataset(
    "test_drift_data.csv",
    config={'dim': 2, 'drift_speed': 0.05, 'noise_type': 'pareto'}
)

results = benchmark.evaluate(
    optimizer_class=MySGD,
    optim_params={'lr': 0.05},
    phases=['heatmap', 'stress']
)

viz = BenchmarkVisualizer(results)
viz.show_dashboard()
```
